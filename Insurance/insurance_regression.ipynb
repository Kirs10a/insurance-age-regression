***Insurance Regression***

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
%matplotlib inline
```

```python
# Import data
data = pd.read_csv('insurance.csv') 
# Displaying the first few rows of the dataset to understand its 
# structure
data.head()
```

```python
# Scatter plot of age on x-axis vs charges on y-axis
# Age is in the first column (index 0) and charges are in the last 
# column (index -1)

plt.figure(figsize=(8, 6))
x = data.iloc[:, 0].values
y = data.iloc[:,-1].values

plt.scatter(x,y,color = 'b')
plt.title('Scatter Plot of Age vs Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.show()
```

The scatter plot above shows the relationship between age and insurance charges. There seems to be a general upward trend, indicating that insurance charges may increase with age, although there is significant variation in the data.

```python

# Reshaping the data for sklearn
X = data[['age']]  # Independent variable
y = data['charges']  # Dependent variable

# Creating and fitting the linear regression model
model = LinearRegression()
model.fit(X, y)

# Making predictions
predictions = model.predict(X)

# Plotting the scatter plot and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['charges'], alpha=0.5, label='Data')
plt.plot(data['age'], predictions, color='red', label='Best Fit Line')
plt.title('Age vs Insurance Charges with Best Fit Line')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.legend()
plt.show()
```

The line shows a positive slope, indicating that charges generally increase with age, the spread of the data suggests that age alone might not be sufficient to fully predict insurance charges.
